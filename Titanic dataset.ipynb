```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby(['pclass'])[['survived']].sum()
```

```python
titanic.groupby(['pclass','sex'])[['survived','fare']].sum()
```

```python
titanic.groupby(['pclass','sex'])[['survived','fare']].mean()
```

```python
titanic.groupby(['pclass','sex'],as_index=False)[['survived','fare']].sum()
```

```python
import matplotlib.pyplot as plt
sns.displot(titanic['age'])
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x="survived", y="age", data=titanic)
plt.show()
```

```python
titanic.groupby(['pclass'])[['survived','fare']].agg(['sum','mean'])
```

```python
titanic.groupby(['sex','class'])['survived'].addregate('mean')
```

```python
titanic.pivot_table('survived', index ='sex', columns='class')
```

```python
agegroup = pd.cut(titanic['age'],[0,18,80])
df_pivot = titanic.pivot_table('survived', index =['sex',agegroup],columns='class')
df_pivot
```

```python
agegroup = pd.cut(titanic['age'],[0,18,80])
df_pivot = titanic.pivot_table('survived', index =['sex',agegroup],columns='class',aggfunc='sum')
df_pivot
```

```python
fare = pd.qcut(titanic['fare'],2)
titanic.pivot_table('survived', ['sex',agegroup],[fare,'class'], aggfunc='sum')
```
